```python
#buat fungsiuntuk membersihkan data menjadi spesialwords seperti halnya diatas
def spesial_words(title, description):
    title = nfx.remove_stopwords(title)
    title = nfx.remove_special_characters(title)
    title = nfx.remove_puncts(title)
    title = nfx.remove_multiple_spaces(title)
    title = nfx.remove_emojis(title)
    title = title.lower()
    
    description = nfx.remove_stopwords(description)
    description = nfx.remove_special_characters(description)
    description = nfx.remove_puncts(description)
    description = nfx.remove_multiple_spaces(description)
    description = nfx.remove_emojis(description)
    description = description.lower()
    
    spesial_words = title + ' ' + description
    
    return spesial_words

#buat kolom baru dengan nama spesial_words2
df['spesial_words'] = df.apply(lambda x: spesial_words(x['title'], x['description']), axis=1)
```
